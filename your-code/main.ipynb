#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
f = plt.figure() #ALWAYS PUT THIS, else title won't work
f.suptitle('linear scale graphs', size = 15)
y1 = x**2
y2 = np.exp(x)

plt.plot(x, y1, 'brown', x, y2, 'blue')
plt.xlim(90, 100)  

#plt.set_yscale('log')
plt.xlabel('x values', size = 12)
plt.ylabel('y values', size = 12)
plt.show()
```

```python
from matplotlib import pyplot
f = plt.figure() #ALWAYS PUT THIS, else title won't work
f.suptitle('log scale graphs', size = 15)
y1 = x**2
y2 = np.exp(x)

plt.plot(x, y1, 'red', x, y2, 'black')
plt.xlim(0, 50)

pyplot.yscale('log')
plt.xlabel('x values', size = 12)
plt.ylabel('y values', size = 12)
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

```python
# MEDIAN STEPS BY MONTH_ENCODED

xticks = ['Week', 'Weekend']
steps_month = fitbit.groupby(['Months_encoded'])['Steps'].median()
plt.bar(steps_month.index, steps_month.values)
plt.xlabel('Months', size = 12)
plt.ylabel('Median steps', size = 12)
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND

steps_month = fitbit.groupby(['Work_or_Weekend'])['Steps'].median()
plt.bar(steps_month.index, steps_month.values)
plt.xticks(np.arange(2), ('weekend', 'weekday'))
plt.ylabel('Median steps', size = 12)
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
f , (b1,b2,b3) = plt.subplots(nrows = 1, ncols = 3, figsize = (15,5))
for i,j in zip(['Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active'],(b1, b2, b3)):
    x = fitbit['Steps']
    y = fitbit[i]
    j.scatter(x, y)
    j.set_title(i)
    j.set(xlabel = 'Steps', ylabel = i)
plt.show()
```
